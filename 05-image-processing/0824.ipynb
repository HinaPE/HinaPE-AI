```python
import os

os.environ["OPENCV_IO_ENABLE_OPENEXR"] = "1"
import cv2
from matplotlib import pyplot as plt
import numpy as np
```

```python
color = cv2.imread('render/front/Image.front1.0090.exr', cv2.IMREAD_UNCHANGED)
color = color[:, :, :3]
gray = cv2.cvtColor(color, cv2.COLOR_BGR2GRAY)
gray = cv2.normalize(gray, None, 0, 1, cv2.NORM_MINMAX)
gray = (255*gray).astype('uint8')
```

```python
plt.imshow(gray, cmap='gray')
plt.axis('off')
```

```python
canny = cv2.Canny(gray, 50, 60)
plt.imshow(canny, cmap='gray')
plt.axis('off')
```

```python
edge_color = 255
gray_with_edges = gray.copy()
gray_with_edges //= 10
gray_with_edges[canny != 0] = 255
plt.imshow(cv2.cvtColor(gray_with_edges, cv2.COLOR_BGR2RGB))
plt.axis('off')
```

```python
_, binary_image = cv2.threshold(gray, 1, 255, cv2.THRESH_BINARY)
plt.imshow(binary_image, cmap='gray')
plt.axis('off')
```

```python
contour = cv2.Canny(binary_image, 100, 120)
plt.imshow(contour, cmap='gray')
plt.axis('off')
```
